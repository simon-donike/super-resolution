```python
import rasterio,numpy as np
from rasterio.windows import Window
import matplotlib.pyplot as plt
import os

import warnings

warnings.filterwarnings("ignore")
```

```python
path = "/home/simon/CDE_UBS/thesis/data_collection/sen2/date_mosaics/"
file = "merged_reprojected20180722_mosaic.tif"
filepath = path+file
```

```python
files = os.listdir(path)
```

```python
def load_window(path,x,y,size):
    import rasterio
    with rasterio.open(path) as src:
        b1 = src.read(1, window=Window(x, y, size, size))
        b2 = src.read(2, window=Window(x, y, size, size))
        b3 = src.read(3, window=Window(x, y, size, size))
        im = np.dstack((b1,b2,b3))
        del b1,b2,b3


    return(im)
```

```python
for i in files:
    if i[:6] =="merged":
        
        fig, (ax1, ax2) = plt.subplots(1, 2)
        fig.suptitle(i)
        part = load_window(path+i,x=1,y=1,size=4000)
        ax1.imshow(part)
        
        part = load_window(path+i,x=4000,y=4000,size=4000)
        ax2.imshow(part)
        
        plt.show()
```

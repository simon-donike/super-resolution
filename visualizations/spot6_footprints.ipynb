```python
import geopandas
import rasterio
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
```

```python
df = geopandas.read_file("data/spot6_dates.geojson")
```

```python
print(df.columns)
df.plot(column="date")
```

```python
ls = []
for i in df["date"]:
    ls.append(datetime.strptime(i, '%Y%m%d').timetuple().tm_yday)
df["DOY"] = ls
```

```python
df.plot(column="DOY")
```

```python
df.to_file("spot6_DOY.geojson", driver="GeoJSON")
```
